```python
import pandas as pd
import dask.dataframe as dd
import numpy as np
import time
from dask.distributed import Client, LocalCluster

# Ajustar la configuraci√≥n de Dask
import dask
dask.config.set({'distributed.scheduler.work-stealing': False})

# Crear un cl√∫ster local con el n√∫mero especificado de trabajadores
num_workers = 4
cluster = LocalCluster(n_workers=num_workers, threads_per_worker=1)
client = Client(cluster)

# Generar datos de ejemplo
num_files = 5
rows_per_file = 1000000
product_names = ['A', 'B', 'C', 'D']

# Crear y guardar archivos CSV simulados
import os
if not os.path.exists('data'):
    os.makedirs('data')

for i in range(num_files):
    df = pd.DataFrame({
        'product': np.random.choice(product_names, size=rows_per_file),
        'sales': np.random.rand(rows_per_file) * 100
    })
    df.to_csv(f'data/file_{i}.csv', index=False)

# Sin Dask: usar Pandas
start_time = time.time()

# Leer todos los archivos CSV en un solo DataFrame de Pandas
df_pandas = pd.concat([pd.read_csv(f'data/file_{i}.csv') for i in range(num_files)])
total_sales_pandas = df_pandas.groupby('product')['sales'].sum()

end_time = time.time()
print("Pandas - Total Sales:\n", total_sales_pandas)
print("Pandas - Time taken:", end_time - start_time, "seconds")

# Con Dask: usar Dask DataFrame
start_time = time.time()

# Leer todos los archivos CSV en un DataFrame de Dask
df_dask = dd.read_csv('data/file_*.csv')
total_sales_dask = df_dask.groupby('product')['sales'].sum().compute()

end_time = time.time()
print("Dask - Total Sales:\n", total_sales_dask)
print("Dask - Time taken:", end_time - start_time, "seconds")

# Cerrar el cl√∫ster de Dask
client.close()
cluster.close()
```

```python
import os
import xarray as xr
import pandas as pd
from glob import glob

def verificar_y_agregar_hora(name):
    # Abre el dataset
    ds = xr.open_dataset(name)
    
    # Verifica si la variable 'time' tiene horas
    if 'time' in ds.coords:
        tiempo = ds['time'].dt.time
        
        # Si no hay horas (solo fechas), se agrega la hora del nombre del archivo
        if all(t == pd.Timestamp('00:00:00').time() for t in tiempo):
            # Extrae la hora del nombre del archivo
            hora = name.split('_')[3][:2]
            print(f"Hora extra√≠da del archivo: {hora}")
            
            # Convierte a DataFrame para agregar la hora
            df = ds.to_dataframe().reset_index()
            
            # Agrega la hora extra√≠da al campo 'time'
            df['time'] = df['time'].astype(str) + f" {hora}:00"
            
            # Convierte de vuelta a un dataset
            df['time'] = pd.to_datetime(df['time'])
            ds = df.set_index(['time', 'lat', 'lon']).to_xarray()
        else:
            print(f"El dataset ya tiene horas: {tiempo}")
    else:
        print("El dataset no tiene una variable 'time'")
    
    return ds

# Ruta de entrada y salida
input_folder = '/data/users/mandrade/DMARTINEZ/paper2/SeriesBoxLOR1/'
output_folder = '/home/mandrade/DMARTINEZ/paper2_dataordenada/nuevosCalculos/'

# Crear la carpeta de salida si no existe
os.makedirs(output_folder, exist_ok=True)

# Buscar todos los archivos .nc en la carpeta de entrada
nc_files = glob(os.path.join(input_folder, '*.nc'))

# Procesar cada archivo
for nc_file in nc_files:
    # Obtener el nombre del archivo sin la ruta
    filename = os.path.basename(nc_file)
    
    # Verificar y agregar hora si es necesario
    ds = verificar_y_agregar_hora(nc_file)
    
    # Exportar el nuevo dataset a la carpeta de salida
    output_path = os.path.join(output_folder, filename)
    ds.to_netcdf(output_path)
    print(f"Archivo exportado a: {output_path}")
```
